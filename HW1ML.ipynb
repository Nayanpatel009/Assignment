```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Load the training data from CSV
train_data = pd.read_csv('train.csv')

# Load the testing data from CSV
test_data = pd.read_csv('test.csv')


# Extract features (X) and target variable (y) from training data
X_train = train_data[['YearBuilt']]
y_train = train_data['SalePrice']

# Extract features (X) and target variable (y) from testing data
X_test = test_data[['YearBuilt']]
y_test = test_data['SalePrice']

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Visualize the training data, testing data, and the regression line
plt.scatter(X_train, y_train,)
plt.scatter(X_test, y_test, color='red',)
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Linear Regression Line')
plt.xlabel('YearBuilt')
plt.ylabel('SalePrice')
plt.legend()
plt.show()

# Print the model's coefficients and intercept
print(f"Coefficient: {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the training data from CSV
train_data = pd.read_csv('train.csv')

# Load the testing data from CSV
test_data = pd.read_csv('test.csv')

# Extract features (X) and target variable (y) from training data
X_train = train_data[['YearBuilt']]
y_train = train_data['SalePrice']

# Extract features (X) from testing data
X_test = test_data[['YearBuilt']]

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Visualize the training data, testing data, and the regression line
plt.scatter(X_train, y_train, label='Training Data')
plt.plot(X_test, y_pred, color='red', linewidth=3, label='Linear Regression Line')
plt.xlabel('Feature Column')
plt.ylabel('SalePrice')
plt.legend()
plt.show()

# Print the model's coefficients and intercept
print(f"Coefficient: {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")

# Evaluate the model's performance on training data (optional)
y_train_pred = model.predict(X_train)
mse_train = mean_squared_error(y_train, y_train_pred)
print(f"Mean Squared Error on Training Data: {mse_train}")
```
