```python
#problem 2 Count the number of distinct words in your dictionary and The words in the dictionary should be displayed in an alphabetic order
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import string
import re
import string
import collections


datafile = "/content/Shakespeare.txt"
# Read the text from the file
with open(datafile, "r") as f:
    text = f.read()

text = text.translate(str.maketrans('', '', string.punctuation))

# Split the text into words
words = text.lower().split()

# Create a dictionary to count the occurrences of each word
word_counts = collections.Counter(words)

# Get the distinct words and sort them in alphabetical order
distinct_words = sorted(word_counts.keys())

# Count the number of distinct words
number_of_distinct_words = len(distinct_words)

# Print the results
print(f"Number of distinct words: {number_of_distinct_words}")
for word in distinct_words:
    print(word)
```

```python
#problem Select three sentences from the file, then use any POS tagging tools to identify POS tags in the selected sentences.
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# Download NLTK data for POS tagging
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def pos_tagging(sentence):
    # Tokenize the sentence into words
    words = word_tokenize(sentence)

    # Perform POS tagging
    pos_tags = pos_tag(words)

    return pos_tags

# Example sentence
selected_sentence = "And only herald to the gaudy spring and Within thine own bud buriest thy content And tender churl mak'st waste in niggarding"

# Perform POS tagging on the selected sentence
tags = pos_tagging(selected_sentence)

# Display the result
print("Original Sentence:", selected_sentence)
for word in tags:
    print(word)


```
