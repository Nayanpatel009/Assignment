```python
#Problem1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/Nayanpatel009/Assignment/main/VaccineData-0.csv')
df.describe()


# Calculate death rates per 100,000 population
df['Death_Rate'] = (df['Deaths'] / df['Case']) * 100

# Sort the DataFrame by death rate in descending order
df = df.sort_values(by='Death_Rate', ascending=False)

# Plot the data with labels on bars
plt.figure(figsize=(20, 6))
bars = plt.bar(df['Counties'], df['Death_Rate'], color='skyblue')

# Add labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='top', rotation=90)

plt.xlabel('Counties')
plt.ylabel('Death Rate per 1000')
plt.title('Death Rates of Counties in Alabama')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Problem2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/Nayanpatel009/Assignment/main/VaccineData-4.csv')

# Group the vaccine providers by county and count the number of providers in each county
grouped_df = df.groupby('County').size().to_frame(name='count').reset_index()
plt.figure(figsize=(20, 6))
# Create a bar chart
plt.bar(grouped_df['County'], grouped_df['count'])

for i, count in enumerate(grouped_df['count']):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')


plt.xlabel('County')
plt.ylabel('Number of Vaccine Providers')
plt.title('Number of Vaccine Providers by County')
plt.xticks(rotation=90, ha='center')

plt.show()
```

```python
#Problem3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/Nayanpatel009/Assignment/main/VaccineData-2.csv')
df.describe()


# Calculate death rates per 100,000 population
df['Death_Rate'] = (df['Population > 16'] / df['People Completely Vaccinated']) * 100

# Sort the DataFrame by death rate in descending order
df = df.sort_values(by='Death_Rate', ascending=False)

# Plot the data with labels on bars
plt.figure(figsize=(20, 6))
bars = plt.bar(df['County'], df['Death_Rate'], color='skyblue')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='top', rotation=90)

plt.xlabel('Counties')
plt.ylabel('Death Rate per 100')
plt.title('Death Rates of Counties in Alabama')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
